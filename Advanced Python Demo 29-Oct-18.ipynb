#### Pythons decorators allow you to extend and modify the behavior of a functions, methods, and classes without permanently modifying them.
#### A function can take a function as argument (the function to be decorated) and return the same function with or without extension.



#### Decorator types

 - Function Decorators
  - Can define a function inside another function.
  - A function can be passed as parameter to another function.
  - A function can also return another function
 - Method decorators
  - @classmethod - To handle values declared in the class. Cant access values of instances.
  - @staticmethod  Do not have handle to the class and the instances also.
 - Class decorators - to add a member to a class



#### Tracing function calls using Python decorators

 - Lets suppose you want to trace calls to a function. 
 - Every time the function gets called, you want to print/log its name and the values of any arguments passed by the caller. 

### Lets take 2 simple functions which do not have any tracing.

```python
def square(x):
    return x*x

def double(x):
    return x+x
```

```python
## No any tracing of the function calls
print(square(50))
print(double(50))
```

### Decorator definition

### Below is the definition of the decorator function

```python
# Slide 20

# Decorator definition
def trace(f):
    def g(x): 
        print("Being traced - ", f.__name__, x)
        return f(x)
    return g
```

### Below are the function definitions using decorators

### Note that we just need to add @trace at the top of each function

```python
# Slide 21

@trace
def square(x):
    return x*x

@trace
def double(x):
    return x+x
```

### Now we can @trace to each function which is to be traced.

```python
# Slide 21

print(double(50))
```

```python
# Slide 21

print(square(50))
```

### Python Decorators - @staticmethod

```python
# Slide 22

class Student(object): 	 # The class definition
    @staticmethod
    def is_full_name(name_str):  # It do not accept handle to any class.
        names = name_str.split(' ')
        return len(names) > 1
```

```python
# Slide 22

print(Student.is_full_name('Scott Robinson'))   # True  
print(Student.is_full_name('Scott'))            # False 
```

### Python Decorators - @classmethod

```python
# Slide 23

class Kls(object):	# The class definition
    no_inst = 0		# The variable which stores values specific to class.
 
    def __init__(self):
        Kls.no_inst = Kls.no_inst + 1         # Increments the values of the class. Note use of Kls instead of Self
 
    @classmethod		    # It accepts only the class object.
    def get_no_of_instance(cls_obj):    
        return cls_obj.no_inst                    # return the variable value from class object
 
```

```python
# Slide 23

ik1 = Kls()			    # create a class object. __init__ is called.
ik2 = Kls()			    # create another class object. __init__ is called again.
 
print(ik1.get_no_of_instance())  # Calls @classmethod and Prints 2. 2 is the count of class instances
print(Kls.get_no_of_instance())   # Again calls @classmethod and Prints 2 . 2 is the count of class instances
```

### Python @property

#### Pythonic way to use getters and setters.
#### Used to hide the attribute temperature (make it private) and 
#### Define new getter and setter interfaces to manipulate it. 
#### This can be done as follows.

```python
# Slide 25

class Celsius:
    def __init__(self, temperature = 0):
        self._temperature = temperature

    @property			#Its the getter of the property - temperature
    def temperature(self):
        print("Getting value")
        return self._temperature

    @temperature.setter		 #Its the setter of the property  temperature. We can apply validation here
    def temperature(self, value):
        if value < 0:
            raise ValueError("Temperature below 0 is not valid") 
        print("Setting value")
        self._temperature = value
```

```python
# Slide 26
c = Celsius()

c.temperature = 50

print(c.temperature)

print("Setting invalid value")
c.temperature = -10
```
